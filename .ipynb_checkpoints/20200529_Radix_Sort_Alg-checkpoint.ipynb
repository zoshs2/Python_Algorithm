```python
def countingSort1(arr, digit):
    n = len(arr)
  
    # 배열의 크기에 맞는 output 배열을 생성하고 10개의 0을 가진 count란 배열을 생성한다. 
    output = [0] * (n)
    count = [0] * (10) # 0, 1, 2, 3, ..., 9 까지 10개
    
    #digit, 자릿수에 맞는 count에 += 1을 한다. 
    for i in range(n):
        index = int(arr[i]/digit) 
        # print(index)
        count[ (index)%10 ] += 1 # 포인트!!!!  자릿수에 마다 count를 수행한다.
    
    # print(count)
    # count 배열을 수정해 digit으로 잡은 포지션을 설정한다.  
    for i in range(1,10):
        count[i] += count[i-1]   ## 이거 도저히 무슨 의미인지 모르겠다......;;;;;;;;;;;;;;;;
        # print(i, count[i])
    
    # print("다시 count : ", count)
    # 결과 배열, output을 설정한다. 설정된 count 배열에 맞는 부분에 arr원소를 담는다.   
    i = n - 1
    while i >= 0:
        index = int(arr[i]/digit)
        output[ count[ (index)%10 ] - 1] = arr[i]
        count[ (index)%10 ] -= 1
        i -= 1
    
    # print("output : ", output)

    #arr를 결과물에 다시 재할당한다.  
    for i in range(0,len(arr)): 
        arr[i] = output[i]
    
    # print("arr : ", arr)
# Method 
def radixSort1(arr):
    # arr 배열중에서 maxValue를 잡아서 어느 digit, 자릿수까지 반복하면 될지를 정한다. 
    maxValue = max(arr)
    # 자릿수(digit)마다 countingSorting을 시작한다. 
    digit = 1 # 일의 자리부터 시작 
    while int(maxValue/digit) > 0: # int 라서~~ 0으로 해도 무한루프 안빠지는고만. 1 미만인 소수로 떨어지면 어차피 0이니까.
        countingSort1(arr,digit)
        digit *= 10 # 십의 자리, 백의 자리 쭉쭉쭉 늘려나가며 counting 해나감.
        
```

```python
# Radix Sort Algorithm ver.2
def countingSort(nums, digit_index, base):
    # 먼저 base만큼의 list들의 list를 생성해 줍니다.
    # 예를 들면 주어진 수들이 해당 자리수에 어떤 숫자들이 있느냐에 따라 각 리스트로 분배될 것입니다.
    buckets = [[] for _ in range(base)] # 
     
    for n in nums:
        # 각 숫자마다 돌면서 digit_index번째 자리수를 구하고, bucket에 집어넣습니다.
        digit = int((n / (base**digit_index)) % base) # 자릿수마다 값을 뽑아내는 알고리즘 **
        buckets[digit].append(n)
     
    # 최종적으로 그 bucket을 순차적으로 돌면서 결과 리스트에 넣으면 됩니다.
    result = []
    for b in buckets:
        for n in b: # 큐(Queue) 동작방식처럼 First In First Out(FIFO)로 행동한다.
            result.append(n)
     
    return result # 자릿수별로 재배열.

def radixSort(nums):
    # 가장 먼저 최대값이 얼마인지를 찾습니다. 최대값을 기준으로!
    maximum = max(nums)
    max_n_digits = len(str(maximum)) # 값이 몇자리 수인지 max_n_digits에 기록. (ex. 1532이라면 max_n_digits 은 4가 된다.)
     
    # 1의 자리부터 각 자리마다 counting sort를 시행합니다.
    result = nums
    for i in range(max_n_digits):
        result = countingSort(result, i, 10)
         
    return result
```

```python
def get_digit(number, d, base): # 일의 자리 => d = 0, 십의 자리 => d = 1, 백의 자리 => d = 2 ....
    return (number // base ** d) % base ### 우선 순위 : // << ** !!!!!!!!!
```
