```python
# 현존하는 가장 빠른 정렬 알고리즘.
# 2개의 그룹으로 나눠서 각각의 영역에서 따로 정렬을 하는 방식
# Pivot : 기준이 되는 수
# Pivot보다 작은 영역과 Pivot보다 큰 영역으로 계속 나눠간다.
# 부분리스트의 크기가 1 또는 0 (<=1) 일 때 정렬수행을 중단한다.

# 퀵 정렬 알고리즘 ver.1 : 부분 리스트를 나누어 퀵 정렬을 수행할 때마다 리스트를 계속해서 만들기 때문에 메모리 사용에 비효율적이라 할 수 있다.
# 코드 자체는 상당히 직관적. 다만 입력데이터 크기가 클때는 비효율적인 메모리 사용
def Quick_Sort(my_list):
    if len(my_list) <= 1:
        return my_list # 리스트크기가 1이하면 빠꾸
    pivot = my_list[len(my_list) // 2] # list의 가운데에 위치한 값을 Pivot으로 선정. (근데 이게 제일 큰 최댓값이라면 정말 기분이 좋지 않겠군.)
    lesser_arr, equal_arr, greater_arr = [], [], [] # 부분 빈 리스트들을 만들어준다 (Pivot보다 작은 영역/ Pivot과 같은 영역/ Pivot보다 큰 영역)
    for num in my_list:
        if num < pivot: # Pivot보다 작은 수들은 lesser_arr에 담아야겠지
            lesser_arr.append(num)
        elif num > pivot: # Pivot보다 큰 수들은 greater_arr에 담아야겠지
            greater_arr.append(num)
        else: # 위와 같은 조건 이외의 상황은 같은 경우밖에 없겠지. 이땐 equal_arr에 담아야할거야
            equal_arr.append(num)
        
    # 1단계 분할이 끝났으면
    return Quick_Sort(lesser_arr) + equal_arr + Quick_Sort(greater_arr) # 재귀호출을 통해서 나눠진 Pivot보다 작은 영역과 큰 영역안에서 다시 Pivot을 정하고
                                                                        # Quick_Sort를 수행한다. 
                                                                        # 이전에 말했지만, 이 재귀는 부분 리스트의 크기가 1이하일때까지 계속 반복된다.
            
```

```python
import random
import time

if __name__ == "__main__":
    list = []
    input_n = input("정렬할 초기 리스트의 갯수 : ")
    for i in range(int(input_n)):
        list.append(random.randint(1, 10000))
    
    print("<<정렬 전>>")
    print(list)
    
    print("<<정렬 후>>")
    start_time = time.time()
    list = Quick_Sort(list)
    print(list)
    running_time = time.time() - start_time
    print("소요 시간 : {}".format(running_time))
```
