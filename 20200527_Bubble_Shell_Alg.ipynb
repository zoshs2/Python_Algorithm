```python
# 거품 정렬 알고리즘 (Bubble Sort Algorithm) : 순차적으로 '바로 옆에 있는 데이터와 비교'해서 옆의 데이터가 크면 자신과 위치를 변경한다.
import random
import time

def bubble_sort(random_list):
    for start_index in range(len(random_list) - 1):
        for index in range(1, len(random_list) - start_index):
            if random_list[index-1] > random_list[index]:
                print("1 : ", random_list)
                temp = random_list[index-1]
                random_list[index-1] = random_list[index]
                random_list[index] = temp
                print("2 : ", random_list)
        print("===========================================")
```

```python
# 셸 정렬 알고리즘 (Shell Sort Algorithm) : 기본구조는 삽입 정렬 알고리즘과 비슷.
#                                       1. 정렬할 데이터를 일정 구간별로 쪼갠다.
#                                       2. 구간 내에서 정렬을 한다.
#                                       3. 정렬이 끝나면, 구간들을 합쳐서 한번더 정렬한다.
#                                       -> 성능면에서 삽입 정렬 알고리즘과 비교도 되지 않을 정도로 우수하다.
import random
import time

def shell_sort(random_list):
    h = 1
    while h < len(random_list):
        h = h * 3 + 1
    h = h // 3 # // 는 몫을 반환한다. (%는 나머지)
    
    while h > 0:
        for i in range(h):
            start_index = i + h
            
            while start_index < len(random_list):
                temp = random_list[start_index]
                insert_index = start_index
                
                while insert_index > h-1 and random_list[insert_index-h] > temp:
                    random_list[insert_index] = random_list[insert_index-h]
                    insert_index = insert_index - h
                
                random_list[insert_index] = temp
                start_index = start_index + h
                
            h = h // 3
```

```python
# 위에꺼 무한루프에 자꾸 무한루프에 빠짐 ㅡ ㅡ
# 셸 정렬 알고리즘 ver.2
def shellSort(alist):
    sublistcount = len(alist)//2
    while sublistcount > 0:

        for startposition in range(sublistcount):
            gapInsertionSort(alist,startposition,sublistcount)

        print("After increments of size",sublistcount,
                                   "The list is",alist)

        sublistcount = sublistcount // 2

def gapInsertionSort(alist,start,gap):
    for i in range(start+gap,len(alist),gap):

        currentvalue = alist[i]
        position = i

        while position>=gap and alist[position-gap]>currentvalue:
            alist[position]=alist[position-gap]
            position = position-gap # while position >= gap 조건 떄문에, position은 음수를 가져봤자야
                                    # list[음수]로 떨어져서 >currentvalue와 비교되는 걱정은 할 필요없어.

        alist[position]=currentvalue

```
