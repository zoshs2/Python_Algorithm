```python
# 선택 정렬 알고리즘 : 정렬되지 않은 데이터 중에 '가장 작은 값을 찾아서' 정렬을 하는 방식
import random

def selected_sort(random_list):
    c_cnt = 0
    s_cnt = 0
    for sel in range(len(random_list)-1): # 어차피 마지막 argument는 자동으로 정렬되어 세팅된다. 
                                          # 따라서 전체 리스트 길이에서 마지막을 빼준다는 의미에서 -1을 해서 iteration을 한다.
        min = random_list[sel]
        minindex = sel
        # find min value
        for step in range(sel+1, len(random_list)): # 비교를 하기 위해서는 sel을 포함하지 않는 모든 수를 다 검색해야하므로, 
                                                    # sel+1부터 random_list의 크기까지 iteration을 수행한다.
            if min > random_list[step]:
                s_cnt += 1
                min = random_list[step]
                minindex = step
            c_cnt += 1
        # swap
        random_list[minindex] = random_list[sel] # 찾은 최소값 위치에다가는 비교대상 값을 넣어주고
        random_list[sel] = min # 찾은 최소값을 앞으로 끌고와서 서로 값을 교환하는 작업.
        
    print(f"비교 횟수 : {c_cnt}")
    print(f"교환 횟수 : {s_cnt}")
```

```python
# 중요! 삽입 정렬 알고리즘 : 선택 정렬 알고리즘처럼 가장 작은 값을 찾는 검색과정이 필요없는 정렬 알고리즘이다.
# 오히려 '순차적으로 정렬'하면서 현재의 값을 정렬되어 있는 값들과 비교하여 위치로 '삽입'하는 방식이다.
import random
import time

def insertion_sort(my_list):
    c_cnt = 0
    s_cnt = 0
    my_list.insert(0, -1) # 마지막 while 루프를 빠져나오기 위한 안전장치. 또한 inx_idx 가 0 에 도달했음에도 -1이란 조건문을 통해서 
                          # IndexError : list index out of range 가 뜰 가능성이 농후하다.
                          # 0번째 argument에 -1을 넣음으로써, 0보다 같거나 큰 양수값만 존재하는 리스트의 값들의 비교에서
                          # 무조건적으로 필히 while문을 빠져나올 수 있다.
    for s_idx in range(2, len(my_list)):
        temp = my_list[s_idx]
        ins_idx = s_idx
        c_cnt += 1
        while my_list[ins_idx-1] > temp:
            s_cnt += 1
            my_list[ins_idx] = my_list[ins_idx-1]
            ins_idx = ins_idx-1
            print(my_list)
 
        my_list[ins_idx] = temp
        print("swap!!! : {}".format(my_list))
    del my_list[0]
    print(f"비교 횟수 : {c_cnt}")
    print(f"교환 횟수 : {s_cnt}")
```

```python
def insertionSort(x):
    for size in range(1, len(x)): # 삽입 정렬 알고리즘은 언제나 두 번째 자료부터 시작!
        val = x[size] # 비교를 할 값을 val에다가 고정.
        i = size # 비교를 할 값의 인덱스를 i에다가 담고 앞으로 변화시키면서 비교시작
        while i > 0 and x[i-1] > val: # 그 비교할 값의 앞에 있는 데이터들과 비교값 val보다 큰지? (앞에 있는 값이 크면 뒤로 보내야하니까)
                                      # 그리고 비교범위는 무조건 i가 0보다는 커야한다. 0에 도달하면, 앞의 데이터가 out of range Erro가 발생한다. (-1 로)
            x[i] = x[i-1] # 앞에 있는 값이 크다면, 바로 뒤에 있는 데이터 위치로 밀어줘야한다.
            i -= 1 # 더 앞에 있는 데이터들도 비교하기 위해서 i를 한칸 감소.
                   # 즉, 비교하는 대상의 val값보다 작고 - () - 큰 값의 범위를 찾고 그 사이에다가 낑겨 넣는겨(삽입!). 
                   # 비교대상은 언제나 val ! 
        x[i] = val 
```

```python
def my_insert_sort(my_list):
    for size in range(1, len(my_list)):
        val = my_list[size]
        i = size
        while i > 0 and my_list[i-1] > val:
            my_list[i] = my_list[i-1]
            i -= 1
        
        my_list[i] = val # 끼워넣을 범위(공간)을 찾았으면 이제 그 자리에다가 val을 삽입!
```

```python
if __name__ == "__main__":
    list = []
    input_n = input("정렬할 데이터의 수 : ")
    for i in range(int(input_n)):
        list.append(random.randint(1, int(input_n)))
    
    print("<정렬 전>")
    print(list)
    
    start_time = time.time()
    insertion_sort(list)
    running_time = time.time() - start_time
    print("<정렬 후>")
    print(list)
    
    print(f"데이터의 크기 : {int(input_n)}")
    print(f"실행 시간 : {running_time}")
```
